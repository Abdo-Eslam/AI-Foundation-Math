```python
x = np.linspace(1 , 10 , 100)
y = 1/x**2 * np.sin(x)
dydx = np.gradient(y , x)
y_int = np.cumsum(y) * (x[1] - x[0])
```

```python
plt.plot(x , y)
plt.plot(x , dydx)
plt.plot(x , y_int)
```

## Examples

```python
x = np.linspace(0, 10, 100)
y = np.exp(-x/10) * np.sin(x)
```

```python
plt.plot(x, y)
```

```python
Mask = (x>=4) & (x<=7)
y_range = y[Mask]
np.mean(y_range)
np.std(y_range)
```

```python
np.percentile(y_range, 80)
```

```python
dydx = np.gradient(y, x)
plt.plot(x, dydx)
```

```python
equal_0 = np.abs(dydx) < 1e-4
dydx[equal_0]
```

```python
plt.plot(x, equal_0)
```
